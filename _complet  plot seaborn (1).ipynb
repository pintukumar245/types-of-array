```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pintu=sns.load_dataset("penguins").head(100)
pintu
```

# Line plot

```python
m={"Male":"*","Female":"s"}
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=pintu,hue="sex",sizes=(140,130),style="sex",markers=m,alpha=1)
plt.grid()
plt.show()
```

# Bar Plot

```python
sns.barplot(x="bill_depth_mm",y="flipper_length_mm",data=pintu,hue="sex",capsize=0.9,errcolor='b')
plt.grid()
plt.show()
```

# Histogram_Plot

```python
c=sns.set_palette("muted")
sns.displot(x="bill_length_mm",y="bill_depth_mm",data=pintu,hue="sex",palette=c,bins=[123,150],log_scale=True)
plt.show()
```

```python
sns.displot(x="flipper_length_mm",data=pintu,bins=[175,180,185,190,195,200,205],rug=True,kde=True)
plt.show()
```

# Scatter Plot

```python
 m={"Male":"*","Female":"s"}

sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=pintu,hue="sex",style="sex",size="sex",sizes=(100,150),markers=m)
plt.show()
```

# Heat Map Plot

```python
import numpy as np
```

```python
var=np.linspace(1,20,20).reshape(4,5)
var
```

```python
sns.heatmap(var)
plt.show()
```

```python
m=pintu.drop(columns=["species",	"island","sex"]).head(5)
m
```

```python
sns.heatmap(m,vmin=1000,vmax=13200,annot=True,linewidth=10,linecolor="red",cbar=False)
plt.show()
```

# Count Plot

```python
sns.countplot(x="island",data=pintu,hue="sex")
plt.show()
```

# violin plot

```python
sns.violinplot(x="body_mass_g",data=pintu)
plt.show()
```

```python
sns.violinplot(x="island",y="body_mass_g",data=pintu,hue="sex",linewidth=1,linecolor="black",inner= "quart")
plt.grid()
plt.show()
```

```python
sns.violinplot(x="island",y="body_mass_g",data=pintu,hue="sex",linewidth=1,linecolor="black",split=True,inner="quart")
plt.grid()
plt.show()
```

# Pair Plot foe only numerical values

```python
k=pintu.drop(columns=["species",	"island"])
k
```

```python
sns.pairplot(k)
plt.show()
```

```python
sns.pairplot(k,hue="sex") 
plt.show()
```

```python

sns.pairplot(k,vars=["bill_length_mm","bill_depth_mm"],hue="sex",kind="reg",markers=["*","o"])
plt.show()
```

```python
sns.pairplot(k,vars=["bill_length_mm","bill_depth_mm"],hue="sex",kind="kde") # kind=scatter,kde,hist,reg
plt.show()
```

# Strip Plot

```python
#island	bill_length_mm	bill_depth_mm	
# m5={"Male":"*","Female":"o"}
sns.stripplot(x="bill_length_mm",y="bill_depth_mm",data=pintu,hue="sex",size=10,marker="*",alpha=0.9)
plt.show()
```

# Box Plot

```python
sns.boxplot(x="bill_depth_mm",data=pintu,hue="sex",showmeans=True,meanprops={"marker":"+","markeredgecolor":"red"})
plt.show()
```

```python
sns.boxplot(x="bill_length_mm",y="bill_depth_mm",data=pintu,hue="sex",showmeans=True,meanprops={"marker":"+","markeredgecolor":"red"},
           linewidth=1)
plt.show()
```

# Factor Plot / Cat Plot

```python
 # m={"Male":"o","Female":"s"}
sns.catplot(x="bill_length_mm",y="bill_depth_mm",data=pintu,hue="sex",marker="s")
plt.show()
```

```python
sns.catplot(x="bill_length_mm",y="bill_depth_mm",data=pintu,hue="sex",kind="bar")
plt.show()
```
